```python
# Quick fix for Jupyter notebooks
import subprocess
import threading
import time
import socket
import sys

def check_service(port=8188, timeout=1):
    """Check if service is running on port"""
    try:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(timeout)
        result = sock.connect_ex(('127.0.0.1', port))
        sock.close()
        return result == 0
    except:
        return False

def start_tunnel(port=8188):
    """Start cloudflared tunnel when service is ready"""
    print(f"Monitoring port {port} for service...")

    # Wait for service to be ready
    while not check_service(port):
        print(".", end="", flush=True)
        time.sleep(1)

    print(f"\n✅ Service detected on port {port}!")
    print("🚀 Starting cloudflared tunnel...")

    try:
        # Start cloudflared with better error handling
        process = subprocess.Popen(
            ["cloudflared", "tunnel", "--url", f"http://127.0.0.1:{port}"],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            universal_newlines=True
        )

        # Monitor output for tunnel URL
        for line in process.stderr:
            print(f"Cloudflared: {line.strip()}")
            if "trycloudflare.com" in line:
                url_start = line.find("https://")
                if url_start != -1:
                    url = line[url_start:].strip()
                    print(f"\n🌐 ComfyUI URL: {url}")
                    break

    except FileNotFoundError:
        print("❌ Cloudflared not found. Installing...")
        # Install cloudflared
        try:
            subprocess.run(["wget", "-O", "/tmp/cloudflared.deb",
                          "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb"],
                          check=True)
            subprocess.run(["sudo", "dpkg", "-i", "/tmp/cloudflared.deb"], check=True)
            print("✅ Cloudflared installed, restarting tunnel...")
            start_tunnel(port)  # Retry
        except Exception as e:
            print(f"❌ Failed to install cloudflared: {e}")
    except Exception as e:
        print(f"❌ Error starting tunnel: {e}")

# Start the tunnel monitor in background
tunnel_thread = threading.Thread(target=start_tunnel, args=(8188,), daemon=True)
tunnel_thread.start()

print("🎯 Tunnel monitor started!")
print("Now run ComfyUI with: !python main.py --dont-print-server")
```

```python
# ComfyUI Tunnel Debugger & Fixer
import subprocess
import threading
import time
import socket
import os
import sys
import signal
from pathlib import Path

class ComfyUITunnelManager:
    def __init__(self, port=8188):
        self.port = port
        self.tunnel_process = None
        self.tunnel_url = None
        self.comfyui_process = None

    def check_port(self, timeout=1):
        """Check if ComfyUI is running on the specified port"""
        try:
            sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            sock.settimeout(timeout)
            result = sock.connect_ex(('127.0.0.1', self.port))
            sock.close()
            return result == 0
        except Exception as e:
            print(f"Port check error: {e}")
            return False

    def install_cloudflared(self):
        """Install cloudflared with comprehensive error handling"""
        print("🔧 Installing cloudflared...")

        try:
            # Method 1: Try package manager first (fastest)
            result = subprocess.run(
                ["curl", "-L", "--fail", "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64",
                 "-o", "/tmp/cloudflared"],
                capture_output=True, text=True, timeout=60
            )

            if result.returncode == 0:
                os.chmod("/tmp/cloudflared", 0o755)
                subprocess.run(["sudo", "mv", "/tmp/cloudflared", "/usr/local/bin/"], check=True)
                print("✅ Cloudflared installed via direct download")
                return True

        except Exception as e:
            print(f"Direct download failed: {e}")

        try:
            # Method 2: Traditional deb package
            print("📦 Trying deb package installation...")

            # Download
            download_result = subprocess.run([
                "wget", "-q", "--timeout=30", "--tries=3",
                "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb",
                "-O", "/tmp/cloudflared.deb"
            ], capture_output=True, text=True)

            if download_result.returncode != 0:
                print(f"Download failed: {download_result.stderr}")
                return False

            # Install
            install_result = subprocess.run([
                "sudo", "dpkg", "-i", "/tmp/cloudflared.deb"
            ], capture_output=True, text=True)

            if install_result.returncode != 0:
                print("Fixing dependencies...")
                subprocess.run(["sudo", "apt-get", "install", "-f", "-y"], capture_output=True)
                # Try install again
                install_result = subprocess.run([
                    "sudo", "dpkg", "-i", "/tmp/cloudflared.deb"
                ], capture_output=True, text=True)

            if install_result.returncode == 0:
                print("✅ Cloudflared installed via deb package")
                return True
            else:
                print(f"Deb installation failed: {install_result.stderr}")
                return False

        except Exception as e:
            print(f"Deb installation error: {e}")
            return False

    def check_cloudflared(self):
        """Check if cloudflared is available and working"""
        try:
            result = subprocess.run(["cloudflared", "--version"],
                                  capture_output=True, text=True, timeout=10)
            if result.returncode == 0:
                print(f"✅ Cloudflared found: {result.stdout.strip()}")
                return True
            else:
                print(f"❌ Cloudflared check failed: {result.stderr}")
                return False
        except FileNotFoundError:
            print("❌ Cloudflared not found in PATH")
            return False
        except Exception as e:
            print(f"❌ Cloudflared check error: {e}")
            return False

    def start_tunnel_with_debug(self):
        """Start cloudflared tunnel with extensive debugging"""
        print(f"🚀 Starting tunnel for port {self.port}...")

        try:
            # Start cloudflared with verbose output
            cmd = ["cloudflared", "tunnel", "--url", f"http://127.0.0.1:{self.port}"]
            print(f"Running command: {' '.join(cmd)}")

            self.tunnel_process = subprocess.Popen(
                cmd,
                stdout=subprocess.PIPE,
                stderr=subprocess.PIPE,
                universal_newlines=True,
                bufsize=1
            )

            print("Cloudflared process started, waiting for tunnel URL...")

            # Monitor both stdout and stderr
            def read_output(stream, name):
                for line in stream:
                    line = line.strip()
                    if line:
                        print(f"[{name}] {line}")

                        # Look for tunnel URL in various formats
                        if "trycloudflare.com" in line:
                            # Extract URL
                            import re
                            url_match = re.search(r'https://[a-zA-Z0-9-]+\.trycloudflare\.com', line)
                            if url_match:
                                self.tunnel_url = url_match.group(0)
                                print(f"\n🌐 TUNNEL URL FOUND: {self.tunnel_url}")
                                print(f"🎯 Access ComfyUI at: {self.tunnel_url}")
                                return

            # Start threads to read both streams
            stdout_thread = threading.Thread(target=read_output, args=(self.tunnel_process.stdout, "STDOUT"), daemon=True)
            stderr_thread = threading.Thread(target=read_output, args=(self.tunnel_process.stderr, "STDERR"), daemon=True)

            stdout_thread.start()
            stderr_thread.start()

            # Wait a bit for the tunnel to establish
            time.sleep(10)

            if self.tunnel_process.poll() is None:
                print("✅ Tunnel process is running")
                if not self.tunnel_url:
                    print("⚠️  Tunnel URL not detected yet, but process is running")
                    print("Check the output above for the URL")
                return True
            else:
                print(f"❌ Tunnel process exited with code: {self.tunnel_process.returncode}")
                return False

        except Exception as e:
            print(f"❌ Error starting tunnel: {e}")
            return False

    def wait_for_service(self, timeout=300):
        """Wait for ComfyUI to be ready"""
        print(f"⏳ Waiting for ComfyUI on port {self.port}...")

        start_time = time.time()
        while time.time() - start_time < timeout:
            if self.check_port():
                print(f"✅ ComfyUI is ready on port {self.port}!")
                return True
            print(".", end="", flush=True)
            time.sleep(2)

        print(f"\n❌ ComfyUI not ready after {timeout} seconds")
        return False

    def setup_complete_tunnel(self):
        """Complete tunnel setup with all debugging"""
        print("🎯 ComfyUI Tunnel Setup")
        print("=" * 50)

        # Step 1: Check/Install cloudflared
        if not self.check_cloudflared():
            print("Installing cloudflared...")
            if not self.install_cloudflared():
                print("❌ Failed to install cloudflared")
                return False

        # Step 2: Wait for ComfyUI
        if not self.wait_for_service():
            print("❌ ComfyUI not available")
            return False

        # Step 3: Start tunnel
        if not self.start_tunnel_with_debug():
            print("❌ Failed to start tunnel")
            return False

        print("✅ Setup complete!")
        return True

    def cleanup(self):
        """Clean up processes"""
        if self.tunnel_process and self.tunnel_process.poll() is None:
            print("🧹 Cleaning up tunnel process...")
            self.tunnel_process.terminate()
            time.sleep(2)
            if self.tunnel_process.poll() is None:
                self.tunnel_process.kill()

# Create global instance
tunnel_manager = ComfyUITunnelManager()

def setup_tunnel_background():
    """Setup tunnel in background thread"""
    try:
        tunnel_manager.setup_complete_tunnel()
    except Exception as e:
        print(f"Background setup error: {e}")

def debug_current_state():
    """Debug the current state of services"""
    print("🔍 DEBUGGING CURRENT STATE")
    print("=" * 40)

    # Check port
    port_open = tunnel_manager.check_port()
    print(f"Port 8188 status: {'✅ OPEN' if port_open else '❌ CLOSED'}")

    # Check cloudflared
    cf_available = tunnel_manager.check_cloudflared()
    print(f"Cloudflared status: {'✅ AVAILABLE' if cf_available else '❌ NOT FOUND'}")

    # Check processes
    try:
        ps_result = subprocess.run(["ps", "aux"], capture_output=True, text=True)
        comfyui_running = "main.py" in ps_result.stdout
        tunnel_running = "cloudflared" in ps_result.stdout

        print(f"ComfyUI process: {'✅ RUNNING' if comfyui_running else '❌ NOT RUNNING'}")
        print(f"Tunnel process: {'✅ RUNNING' if tunnel_running else '❌ NOT RUNNING'}")

    except Exception as e:
        print(f"Process check error: {e}")

    print("\n🔧 RECOMMENDED ACTIONS:")
    if not port_open:
        print("1. Start ComfyUI first: !python main.py --dont-print-server")
    if not cf_available:
        print("2. Install cloudflared: tunnel_manager.install_cloudflared()")
    if port_open and cf_available:
        print("3. Start tunnel: tunnel_manager.start_tunnel_with_debug()")

# Auto-start background monitoring
print("🚀 Starting ComfyUI tunnel monitor...")
monitor_thread = threading.Thread(target=setup_tunnel_background, daemon=True)
monitor_thread.start()

# Provide debug function
print("\n💡 USAGE:")
print("- Run: debug_current_state() to see what's happening")
print("- Run: tunnel_manager.setup_complete_tunnel() to try again")
print("- The monitor is running in background")
print("\nNow start ComfyUI with: !python main.py --dont-print-server")
```
